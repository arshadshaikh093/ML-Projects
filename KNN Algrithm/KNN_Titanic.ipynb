**KNN to classify the Passenger class of Passenger in Titanic Dataset**

```
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import neighbors
```

```
#Mount Google Drive

from google.colab import drive
drive.mount('/content/drive')
```

```
#Import Training Dataset

titanic = pd.read_csv('/content/drive/My Drive/ML and AI/Assignments/Dataset/Titanic Dataset/train.csv')
titanic.head()
```

```
#Check for null values

titanic.isna().sum()
```

```
# check the columns in the dataset

titanic.columns
```

```
#Preprocess the data, Convert the String columns in to numbers

label_encoder = preprocessing.LabelEncoder()
titanic['Sex'] = label_encoder.fit_transform(titanic['Sex'])
titanic['Embarked'] = label_encoder.fit_transform(titanic['Embarked'])
```

```
#Create X and Y datatset
#Drop PassengerId, Name, Cabin, Ticket from dataset as they are not significant

Y = titanic[['Pclass']]
X = titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Pclass'], axis = 1)
X.columns
```

```
#Split the dataset into training and testing dataset

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```
#Check the shapes and sizes to understand and verify that the splits has been done in 30-70% ratio

print("Dataset Shape", titanic.shape)
print("X_train Shape", X_train.shape)
print("X_test Shape", X_test.shape)
print("Y_train Shape", Y_train.shape)
print("Y_test Shape", Y_test.shape)
```

```
#Build the model with k (neighbors) = 3 and predict for test dataset

knn = neighbors.KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
ypred = knn.predict(X_test)
```

```
#Print the accuracy score and confusion matrix

print('Accuracy Score is : ', accuracy_score(Y_test, ypred, normalize=True))
confusion_matrix(Y_test, ypred)
```

```
#Build the model with k (neighbors) = 4 and predict for test dataset

knn = neighbors.KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, Y_train)
ypred = knn.predict(X_test)
```

```
#Print the accuracy score and confusion matrix

print('Accuracy Score is : ', accuracy_score(Y_test, ypred, normalize=True))
confusion_matrix(Y_test, ypred)
```

```
#Build the model with k (neighbors) = 2 and predict for test dataset

knn = neighbors.KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train, Y_train)
ypred = knn.predict(X_test)
```

```
#Print the accuracy score and confusion matrix

print('Accuracy Score is : ', accuracy_score(Y_test, ypred, normalize=True))
confusion_matrix(Y_test, ypred)
```
